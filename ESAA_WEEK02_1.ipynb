## 02 Matplotlib

###맷플롯립의 pyplot 모듈의 이해

matplotlib.pyplot 모듈을 plt라는 이름으로 임포트  
* plot(): 선 그래프 그림  
* title(): 입력값으로 타이틀 문자열을 받아 차트 타이틀 설정  
* show(): 축을 포함한 선 그래프 화면에 나타냄  

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,6])
plt.title("Hello plot")
plt.show()
```

### pyplot의두가지 중요 요소- Figure와 Axes의 이해

* Figure객체: 그림판 크기 조절, 플롯을 최종적으로 화면에 나타내는 역할을 수행, 그림 그리는 역할X  
* Axes: 실제 그림을 그리는 매서드, 대부분의 시각화 수행

1. plt.plot(): 기본 설정된 Axes에서 Axes.plot()함수 호출하여 그림
2. plt.title():내부적으로 다시 Axes.set_title()함수 호출하여 그림

###Figure와 Axes 활용
1. Figure: figsize=(가로크기, 세로크기)인자로 전체 그림판의 크기 조절

```python
plt.figure(figsize=(10,4))
plt.plot([1,2,3],[2,4,6])
plt.title("Hello plot")
plt.show()
```

```python
#figure함수가 Figure객체를 반환함을 확인
figure=plt.figure(figsize=(10,4))
print(type(figure))
```

반환된 Figure은 그림그리는 역할X, 그림판 배경 색 등 설정 가능

```python
plt.figure(figsize=(4,4), facecolor='yellow')
plt.plot([1,2,3],[2,4,6])
plt.title("Hello plot")
plt.show()
```

2. Axes

```python
#plt.axes()를 통해 axes객체 반환
axes=plt.axes()
print(type(axes))
```

별도의 plt.show()가 없더라도 Axes가 할당되어있으면 자동으로 plt.show()와 같은 역할 하는 함수 호출해 그림 표출


3. Figure& Axes 객체 함께 가져오기: plt.subplots()

```python
fig, ax=plt.subplots()
print('fig type: ',type(fig),'\nax type: ',type(ax))
```

```python
#pyplot모듈의 plot()이나 title()이 아닌 axes객체를 바로 이용하기
fig, ax=plt.subplots(figsize=(4,4))
ax.plot([1,2,3],[2,4,6])
ax.set_title('Hello plot')
plt.show()
```

###여러 개의 plot을 가지는 subplot들을 생성하기

* plt.subplots(): 하나의 Figure 상에서 여러 개의 subplot
  > 개별 subplot은 하나의 Axes가짐  
  >하나의 Figure와 여러 Axes 생성하고 반환
  * nrows: subplot들의 배치를 2차원 행렬 형태로 표현할 때 행의 개수
  * ncols: 열의 개수
  * figsize: 모든 subplot들을 포함한 전체 Figure 크기를 설정

```python
fig, (ax1, ax2)=plt.subplots(nrows=1, ncols=2, figsize=(6,3)) #반환되는 Axes객체는 튜플 형태
```

###pyplot의 plot()함수를 이용해 선 그래프 그리기

X좌표값, Y좌표값으로 파이썬 리스트, 넘파이 ndarray, 판다스 DF/Series 모두 가능  
그러나 같은 크기를 가져야 함

```python
import numpy as np
import pandas as pd

x_value=[1,2,3,4]
y_value=[2,4,6,8]
#x_value=np.array([1,2,3,4])
#y_value=np.array([2,4,6,8])
plt.plot(x_value,y_value)
plt.show()
```

color인자를 이용해 선의 색깔 변경

```python
plt.plot(x_value,y_value,color='green')
plt.show()
```

```python
#선 빨간색, 선 형태 대시형, 선 두께 2, 동그라미 형태의 마커, 크기 12
plt.plot(x_value,y_value,color='red',marker='o',markersize=12,linestyle='dashed',linewidth=2)
plt.show()
```

###축 명칭 설정, 축의 눈금값 회전, 범례 설정하기

* 축 명칭:pyplot의 xlabel(), ylabel()

```python
plt.plot(x_value,y_value,color='red',marker='o',markersize=12,linestyle='dashed',linewidth=2)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

* 눈금값 회전: xticks(), yticks()함수 이용

```python
#x좌표값 0-99까지 100개, x축 눈금 90도 회전
x_value=np.arange(0,100)
y_value=x_value*2
plt.plot(x_value,y_value,color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks(rotation=90)
plt.show()
```

맷플롯립은 문자열인 경우 모든 값 나타내지만, 축의 값이 숫자값이면 자동으로 최솟값과 최댓값에 기반해 축의 크기에 따라 눈금 간격을 만들어줌

* xticks의 ticks인자: x축의 눈금 값을 더 세밀하게 나타낼 수 있게 설정

```python
x_value=np.arange(0,100)
y_value=x_value*2
plt.plot(x_value,y_value,color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks(rotation=90,ticks=np.arange(0,100,5))
plt.title('Hello plot')
plt.show()
```

* 범례(legend) 설정: plt.plot() 내에 label인자값 설정

```python
x_value_01=np.arange(0,100)
y_value_01=x_value_01*2
x_value_02=np.arange(0,100)
y_value_02=x_value_02*4

plt.plot(x_value_01,y_value_01,color='green',label='green line')
plt.plot(x_value_02, y_value_02,color='red',label='red line')
plt.xlabel('x axis ')
plt.ylabel('y axis')
plt.legend()
plt.title('Hello plot')
plt.show()
```

* 선과 막대 그래프 함께 그리기

```python
x_value_01=np.arange(0,10)
y_value_01=x_value*2
plt.plot(x_value_01,y_value_01,color='red',marker='o',markersize=12,linestyle='dashed',linewidth=2,label='red line')
plt.bar(x_value_01,y_value_01,color='green',label='bar plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.title('Hello plot')
plt.show()
```

* Axes 객체 메서드를 사용해 시각화 수행: plt.axes()에서 Axes객체 추출
* 변화:
  * plt.plot(),plt.bar() > axes객체의 plot, bar
  * plt.xlabel(), plt.ylabel(), plt.title(): Axes객체의 set_xlabel(), set_ylabel, set_title()
  * plt.legend(): **Axes객체의 legend()** *set안씀*



```python
ax=plt.axes()
ax.plot(x_value_01,y_value_01,color='red',marker='o',linestyle='dashed',label='red line')
ax.bar(x_value_01,y_value_01,color='green',label='bar plot')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.legend()
ax.set_title('Hello plot')
plt.show()
```

###여러 개의 subplots들을 이용해 개별 그래프들을 subplot별로 시각화하기

```python
fig, ax=plt.subplots(nrows=1,ncols=2,figsize=(6,3))
```

반환된 Axes객체들을 가지는 ax변수는 튜플이 아니라 넘파이 ndarray형태  
개별 원소인 ax[0]은 왼쪽, ax[1]은 오른쪽에 생성된 Axes객체를 가르킴

```python
print('ax type: ',type(ax))
print('ax[0] type: ',type(ax[0]))
print('ax[1] type: ', type(ax[1]))
```

<유의사항>  
개별 Axes 객체를 직접 이용해서 시각화 구현해야함

```python
x_value_01=np.arange(0,10)
x_value_02=np.arange(0,20)
y_value_01=x_value_01*2
y_value_02=x_value_02*2

fig, ax=plt.subplots(nrows=1,ncols=2,figsize=(6,3))
ax[0].plot(x_value_01,y_value_01,color='red',marker='o',linestyle='dashed',label='red line')
ax[1].bar(x_value_02,y_value_02,color='green',label='bar plot')
ax[0].set_xlabel('ax[0] x axis')
ax[1].set_xlabel('ax[1] x axis')
ax[0].legend()
ax[1].legend()
ax[0].set_title('Hello line')
ax[1].set_title('Hello bar')
plt.show()
```

nrows또는 ncols가 1인 경우 1차원 배열 인덱싱 방식  
그 이상은 2차원 배열 인덱싱 방식

```python
x_value_01=np.arange(0,10)
x_value_02=np.arange(0,20)
y_value_01=x_value_01*2
y_value_02=x_value_02*2

fig, ax=plt.subplots(nrows=2,ncols=2,figsize=(6,6))
ax[0][0].plot(x_value_01,y_value_01,color='red',marker='o',linestyle='dashed',label='red line')
ax[0][1].bar(x_value_02,y_value_02,color='green',label='green bar')
ax[1][0].plot(x_value_01,y_value_01,color='green',marker='o',linestyle='dashed',label='green line')
ax[1][1].bar(x_value_02,y_value_02,color='red',label='red bar')

ax[0][0].set_xlabel('ax[0][0] x axis')
ax[0][1].set_xlabel('ax[0][1] x axis')
ax[1][0].set_xlabel('ax[1][0] x axis')
ax[1][1].set_xlabel('ax[1][1] x axis')

ax[0][0].legend()
ax[0][1].legend()
ax[1][0].legend()
ax[1][1].legend()

plt.show()
```
