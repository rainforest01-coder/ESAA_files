#2. 코드리뷰
## EDA(탐색적 데이터 분석)
* head(), info()통해 기본적인 데이터 구조 파악
* 성공확률 분포를 hist를 통해 파악
* boxplot을 통해 투자단계에 따른 성공확률 분포 파악
* bar를 이용해 상장여부 및 인수여부에 따른 평균 성공확률 시각화
## 전처리
* object 타입의 기업가치 피처를 해당 범위의 평균값으로 대체하여 실수화
* 결측치 처리:
  * object type: 분야 피처 > fillna('Null')
  * float type: 직원수, 고객수, 기업가치 피처의 결측치를 같은 설립연도, 인수여부, 상장여부를 가진 값들끼리 그룹화하여 평균값으로 채움, 채워지지 않은 결측치의 경우 전체 평균으로 채움
* 파생변수 생성: 연매출을 직원수로 나누어 직원당 연매출과 같은 새로운 피처들을 생성
* 피쳐 엔지니어링: StandardScaler()로 스케일링 및 라벨 인코딩
* 랜덤포레스트로 피처 중요도 확인
##모델링
* 피처 중요도 낮은 피처들 제거 후 KFold n_split=5로 교차검증 및 하이퍼파라미터 튜닝
* scatter로 폴드별 과대 과소 추정 진단
* RandomForestRegressor로 최종 예측값 생성
* 후처리: 예측값에서 상수를 빼 과대추정 보정
* df_test에 ID, target값 저장

#3. 차별점 및 배울점
* 각 피처들을 이용해 다양한 파생변수를 생성
* 피처 중요도를 파악해 상대적으로 불필요한 피처들을 제거
* 상수를 빼 예측값의 과대추정 보정
* 설립연도, 인수여부, 상장여부로 그룹화한 평균값으로 결측치를 채움
