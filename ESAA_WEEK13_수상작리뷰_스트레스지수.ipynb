# 코드리뷰
> 다양한 파생변수를 생성한 뒤 SHAP 분석을 활용하여 임계값을 조정하고 최적의 피처를 선별하였습니다.
모델은 CatBoost가 가장 우수한 성능을 보여 채택하였으며,
과적합 방지를 위해 Seed Ensemble 기법을 적용하였습니다.

## 전처리
* 결측치 다른 값들과 연관이 없는 경우 Unknown을 대체
* 다양한 파생변수 생성: bmi/하루 근무 시간이 11시간을 넘고 불면이 있는 경우 work_sleep_risk변수 등등
* 질병, 교육을 라벨 인코딩
* 피처 중요도가 0에 수렴하거나 제거했을 때 결과가 좋은 피처들 제거
* 조합별로 군집화하여 군집 피처 추가
* 남은 피처들끼리의 곱, 나눗셈, 절댓값 차이, 조화평균 파생변수 생성함수 이용해 피처 생성


## 모델링
* 성능이 가장 좋은 캣부스트 이용
* optuna로 하이퍼 파라미터 최적화
* SHAP임계값보다 피처 중요도가 높은 피처만 고름
* 여러 씨드로 실행 후 가중치 곱해 더함

#차별점 및 배울점
* 다양한 파생변수를 생성한 뒤 SHAP 분석을 활용하여 임계값을 조정하고 최적의 피처를 선별
* 과적합 방지를 위해 Seed Ensemble 기법을 적용
* optuna로 하이퍼 파라미터 최적화
* 피처들끼리의 곱, 나눗셈, 절댓값 차이, 조화평균 파생변수 생성함수 이용해 피처 생성(트리는 “곱(×)”이나 “나눗셈(/)” 같은 복잡한 규칙을 잘 못 표현하므로 도움 될 수 있음)
