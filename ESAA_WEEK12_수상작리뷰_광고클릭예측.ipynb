#코드 리뷰

1. 전처리
* UnderSampling 전략

  - CatBoost 1:1 (3 seed ensemble)
  - XGBoost 10:1 (3 seed ensemble)
  - FiBiNet 10:1 (5 split ensemble)
  > 10:1=neg:pos(클릭여부)로 다운샘플링
*  40:35:25로 앙상블
* 딥러닝 모델 중 FiBiNet을 보조적으로 활용하여 성능 변동을 최소화
* 학습 데이터 3개로 분할
* high_click_numbers.xlsx 생성: seq 안에 등장하는 숫자(number)들이 얼마나 “클릭과 연관이 있는지”를 통계적으로 계산(click_prob = clicked_count / total_count)/한 행에서 중복 제거(클릭 중복 과대반영하지 않기 위함)
* seq를 매핑된 클릭 확률의 평균으로 변환하여 파생변수 생성
* 분할된 데이터 합쳐서 인벤토리 아이디별로 평균과 표준편차 계산

2. 모델링
* Optuna로 하이퍼 파라미터 탐색
* XGBoost 3seed로 예측후 소프트 보팅

#차별점 및 배울점
* seed를 다르게 해서 평균하여 성능을 안정화하고 분산을 줄임
* 데이터가 너무 많은 경우 언더샘플링 활용
*  click_prob = clicked_count / total_count를 만들어 광고별 클릭 확률변수를 생성
* 분할된 데이터를 합쳐 각 인벤토리(광고)별 평균과 표준편차 계산
